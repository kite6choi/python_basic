## User-defined function

`def` 키워드를 사용하여 함수 생성


```python
def 함수이름(입력변수이름):
   출력변수를 만드는 명령
   return 출력변수이름
   
```

숫자 x를 입력하면 두 배가 되도록 하는 함수

````python
def twotimes(x):
    y = 2 * x
    return y
````

함수 사용

````python
twotimes(2)
````

````python
twotimes(10)
````

## 입력이 여러 개인 함수

```python
def 함수이름(입력변수1, 입력변수2, 입력변수3):
   출력변수를 만드는 명령
   return 출력변수이름
```

````python
def add(x, y):
    return x + y
````

````python
add(1, 2)
````

````python
def sum(a, b, c):
    s = a + b + c
    return s
````

````python
sum(1, 2, 3)
````


*   2개의 수를 입력받아 덧셈 뺄셈 곱셈 나눗셈의 결과를 출력하는 함수.
*   2개의 수와 연산자(+,-,*,/)를 입력받아 그 결과를 출력하는 함수
