## `for` 반복문

`for` 반복문(loop)

```
for 카운터변수 in range(반복횟수):
    반복해서 실행할 명령
```

````python
for i in range(10):
    print("=")
````

````python
for i in range(10):
    print(i)
````

````python
for i in range(10):
    print("=", i, "=")
````

````python
for i in range(10):
    print("=",str(i),"=")
````

````python
for i in range(10):
    print("="+str(i)+"=")
````

````python
for i in range(10):
    print("=", str(i), "=", sep='')
````

````python
for i in range(9):
    print("=" + str(i + 1) + "=")
````

<예제>  
누적합 : 1에서 10까지의 합을 구하라.

````python
s=0
for i in range(1,11):
    s = s + i
print("누적합은 = ",s)
````

중첩 반복문(nested loop)

만약 주사위를 두 번 던져서 나온 숫자들의 합이 4의 배수가 되는 경우만 구해야 한다면 다음과 같이 if 조건문을 추가한다.

<구구단 출력>

````python
for i in range(2,10):
    print("= ", i, "단 =")
    for j in range(1,10):
        print(i, "*", j, "=", i*j)
````

## `while` 반복문

```python
while <조건문>:  
    <수행할 문장1>  
    <수행할 문장2>  
    <수행할 문장3>  
```

````python
treeHit = 0
while treeHit < 10:
    treeHit = treeHit +1
    print("나무를 %d번 찍었습니다." % treeHit)
    if treeHit == 10:
        print("나무 넘어갑니다.")
````

<커피 자동 판매기 시뮬레이션>

````python
# 캔커피가 5개 있는 지동판매기 이며, 값은 700원.
coffee = 5
while True:
    money = int(input("돈을 넣어 주세요: "))
    if money == 700:
        print("커피를 줍니다.")
        coffee = coffee -1
    elif money > 700:
        print("거스름돈 %d를 주고 커피를 줍니다." % (money -700))
        coffee = coffee -1
    else:
        print("돈을 다시 돌려주고 커피를 주지 않습니다.")
        print("남은 커피의 양은 %d개 입니다." % coffee)
    if coffee == 0:
        print("커피가 다 떨어졌습니다. 판매를 중지 합니다.")
        break
````
